```python
%%bash
# Write the C code to a file
cat > hello.c << EOF
#include <stdio.h>

int main() {
    printf("Hello, World!\n");
    return 0;
}
EOF

# Compile the C code
gcc hello.c -o hello

# Execute the compiled program
./hello
```
